```python
import pandas as pd
```

```python
messages = pd.read_csv('/content/spam.csv',encoding='latin1')
messages
```

```python
messages.drop(messages[['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4']],axis=1,inplace=True)
```

```python
messages.columns = ['label','message']
```

```python
messages
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i]) # Remove Special Character
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word) for word in review ]
    review = ' '.join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, messages['message']) if i<1]
# i = len,corpus
# Here those 3 record got remove bcoz of this : re.sub('[^a-zA-Z]'
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
#simple_preprocess()
# Convert a Document into List of lowercase tokens, ignoring tokens that are too short or too long
nltk.download('punkt')
```

```python
words = []

for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
def avg_word2vec(doc):
    # remove out-of-vocabulary words
    #sent = [word for word in doc if word in model.wv.index_to_key]
    #print(sent)

    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
                #or [np.zeros(len(model.wv.index_to_key))], axis=0)
```
